```python
import pandas as pd

dataset = pd.read_csv('CarPrice_Assignment.csv')
print(dataset.head())
```

```python
print(dataset.info())
```

```python
print(dataset.isnull().sum())
```

```python
dataset = dataset.dropna()
print("Размер после удаления пропусков:", dataset.shape)
```

```python
numeric_cols = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 
                'enginesize', 'boreratio', 'stroke', 'compressionratio', 
                'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']

for col in numeric_cols:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')

print(dataset.dtypes)
```

```python
cat_cols = ['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 
            'enginelocation', 'enginetype', 'fuelsystem']

for col in cat_cols:
    dataset[col] = pd.factorize(dataset[col])[0]

print(dataset[cat_cols].head())
```

```python
dataset.index = [index for index in range(len(dataset))]
print(dataset.head())
```

```python
dataset = dataset.drop('CarName', axis=1)
print(dataset.columns)
```

```python
for col in numeric_cols:
    min_val = dataset[col].min()
    max_val = dataset[col].max()
    dataset[col] = (dataset[col] - min_val) / (max_val - min_val)

print(dataset[numeric_cols].head())
```
